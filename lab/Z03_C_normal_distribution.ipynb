# Wielowymiarowy rozkład normalny

```python
import matplotlib.pylab as plt
import numpy as np
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.mlab as mlab
from matplotlib import cm
from scipy import stats
from matplotlib.ticker import LinearLocator

```

# Narysujmy gęstość dwuwymiarowego rozkładu Normalnego

```python
mean = np.array([0, 0])
cov = np.array([[1, 0], [0, 1]])
fig = plt.figure()
ax = fig.add_subplot(projection='3d')
x = np.linspace(-3.0, 3.0, 100)
y = np.linspace(-3.0, 3.0, 100)
X, Y = np.meshgrid(x, y)
Z = stats.multivariate_normal(mean, cov).pdf(np.dstack([X, Y]))
ax.plot_surface(X, Y, Z, cmap=cm.coolwarm)
ax.set_zlim(0, 0.2)
plt.show()
```

# Narysujmy kontury gęstości dwuwymiarowego rozkładu Normalnego

```python
x = np.linspace(-3, 3, 100)
y = np.linspace(-3, 3, 100)
X, Y = np.meshgrid(x, y)
mean = np.array([0, 0])
sigma = np.array([[1, 0], [0, 1]])
Z = stats.multivariate_normal(mean, sigma).pdf(np.dstack([X, Y]))
plt.contourf(X, Y, Z, cmap = 'Greens')
plt.colorbar()
plt.show()
```

## Narysujmy poziomice (contour plot) dla mieszanki gaussów

Rozpatrzmy mieszankę gaussów postaci
$$
\frac{1}{2}N(m_1,\Sigma_1)+\frac{1}{2}N(m_2,\Sigma_2).
$$
Proszę narysować poziomice dla przykładu gdy $m_1\neq m_2$, $\Sigma_1=\Sigma_2$

```python
x = np.linspace(-4, 4, 100)
y = np.linspace(-3, 3, 100)
X, Y = np.meshgrid(x, y)
mean_1 = np.array([0, 1])
mean_2 = np.array([1, 0])
sigma_1 = np.array([[1, 0], [0, 1]])
sigma_2 = np.array([[1, 0], [0, 1]])
Z = 0.5 * stats.multivariate_normal(mean_1, sigma_1).pdf(np.dstack([X, Y])) + 0.5 * stats.multivariate_normal(mean_2, sigma_2).pdf(np.dstack([X, Y]))
plt.contourf(X, Y, Z, cmap = 'Greens')
plt.colorbar()
plt.show()
```

# Wylosujmy próbkę z dwuwymiarowego rozkładu Normalnego

Proszę wylosować na dwa sposoby: korzystając z scipy.stats.multivariate_normal oraz numpy.random.multivariate_normal

```python
mean = np.array([0, 0])
cov = np.array([[1, 0], [0, 1]])

X, Y = stats.multivariate_normal(mean, cov).rvs(1000).T
plt.plot(X, Y, '+', markersize = 3, label='scipy')
X, Y = np.random.multivariate_normal(mean, cov, 1000).T
plt.plot(X, Y, '.', markersize = 3, label='numpy')
plt.legend(loc = 0)
plt.show()
```
